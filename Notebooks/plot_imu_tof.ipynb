```python
import pandas as pd

# Load the data from the CSV files
pico_data_path = '/home/luca/raspicar/data/08-10-2024_10-01-54/pico_data.csv'

# Read the CSV files into DataFrames
pico_data_df = pd.read_csv(pico_data_path)

# Display the data to ensure correct loading and review structure
pico_data_df.head()
```

```python
import matplotlib.pyplot as plt

# Plot 2: Gyroscope Data over Time
plt.figure(figsize=(12, 6))
plt.plot(pico_data_df['timestamp'], pico_data_df['gyro_x'], label='Gyro X')
plt.plot(pico_data_df['timestamp'], pico_data_df['gyro_y'], label='Gyro Y')
plt.plot(pico_data_df['timestamp'], pico_data_df['gyro_z'], label='Gyro Z')
plt.title('Gyroscope Data over Time')
plt.xlabel('Timestamp')
plt.ylabel('Gyro Readings')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 3: Accelerometer Data over Time
plt.figure(figsize=(12, 6))
plt.plot(pico_data_df['timestamp'], pico_data_df['accel_x'], label='Accel X')
plt.plot(pico_data_df['timestamp'], pico_data_df['accel_y'], label='Accel Y')
plt.plot(pico_data_df['timestamp'], pico_data_df['accel_z'], label='Accel Z')
plt.title('Accelerometer Data over Time')
plt.xlabel('Timestamp')
plt.ylabel('Accel Readings')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 4: TOF Data over Time
plt.figure(figsize=(10, 5))
plt.plot(pico_data_df['timestamp'], pico_data_df['tof'], label='TOF', color='purple')
plt.title('TOF Data over Time')
plt.xlabel('Timestamp')
plt.ylabel('TOF Reading')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Set up the 3D plot for Accelerometer Data
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(pico_data_df['accel_x'], pico_data_df['accel_y'], pico_data_df['accel_z'], marker='o', linestyle='-', color='b')

# Labeling the axes
ax.set_title('3D Trajectory based on Accelerometer Data')
ax.set_xlabel('Accel X')
ax.set_ylabel('Accel Y')
ax.set_zlabel('Accel Z')
plt.tight_layout()
plt.show()

# Set up the 3D plot for Gyroscope Data
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(pico_data_df['gyro_x'], pico_data_df['gyro_y'], pico_data_df['gyro_z'], marker='o', linestyle='-', color='r')

# Labeling the axes
ax.set_title('3D Trajectory based on Gyroscope Data')
ax.set_xlabel('Gyro X')
ax.set_ylabel('Gyro Y')
ax.set_zlabel('Gyro Z')
plt.tight_layout()
plt.show()
```
